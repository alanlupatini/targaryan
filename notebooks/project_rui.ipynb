```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load data
df = pd.read_csv('../data/analysis_tables/variation_daily_summary.csv')
df['date'] = pd.to_datetime(df['date'])

# Create the full figure with 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(14, 12), sharex=True)
fig.suptitle('Control vs Test: Daily Metrics Over Time', fontsize=16, fontweight='bold')

# Plot 1: daily_unique_visitors
control_vis = df[df['Variation'] == 'Control'].set_index('date')['daily_unique_visitors']
test_vis = df[df['Variation'] == 'Test'].set_index('date')['daily_unique_visitors']
axes[0].plot(control_vis.index, control_vis.values, 'b-o', label='Control', linewidth=2, markersize=4)
axes[0].plot(test_vis.index, test_vis.values, 'r-s', label='Test', linewidth=2, markersize=4)
axes[0].set_ylabel('Unique Visitors')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: daily_unique_visits  
control_visits = df[df['Variation'] == 'Control'].set_index('date')['daily_unique_visits']
test_visits = df[df['Variation'] == 'Test'].set_index('date')['daily_unique_visits']
axes[1].plot(control_visits.index, control_visits.values, 'b-o', label='Control', linewidth=2, markersize=4)
axes[1].plot(test_visits.index, test_visits.values, 'r-s', label='Test', linewidth=2, markersize=4)
axes[1].set_ylabel('Unique Visits')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Plot 3: Daily average visitors (your code fixed)
df['date_only'] = df['date'].dt.date
daily_avg = df.groupby(['date_only', 'Variation']).agg({
    'daily_unique_visitors': 'mean',
    'daily_unique_visits': 'mean'
}).reset_index()
control_avg = daily_avg[daily_avg['Variation'] == 'Control']
test_avg = daily_avg[daily_avg['Variation'] == 'Test']
axes[2].plot(pd.to_datetime(control_avg['date_only']), control_avg['daily_unique_visitors'], 
             'b--', label='Control Visitors', linewidth=2, markersize=6)
axes[2].plot(pd.to_datetime(test_avg['date_only']), test_avg['daily_unique_visitors'], 
             'r--', label='Test Visitors', linewidth=2, markersize=6)
axes[2].set_ylabel('Visitors (Avg)')
axes[2].set_xlabel('Date')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# Demographics (df_demo_clean)

H: The average age of clients engaging with the new process = average age of those with the old process.

H: The average age differs between new vs old process clients.

 Test: Twosample ttest (Welchs) 

```python
print(df_demo_clean.columns)
```

```python
import pandas as pd
from scipy import stats

# Load datasets
df_demo = pd.read_csv("../data/Clean/df_demo_clean.csv")
df_experiment = pd.read_csv("../data/Clean/df_experiment_clean.csv")

# Merge on client_id
df = pd.merge(df_demo, df_experiment, on="client_id")

# Check unique values in variation
print("Process types:", df["variation"].unique())

# Split into two groups: Test vs Control
age_test = df[df["variation"] == "Test"]["client_age"]
age_control = df[df["variation"] == "Control"]["client_age"]

# Perform Welchs t-test
t_stat, p_value = stats.ttest_ind(age_test, age_control, equal_var=False)

# Print results
print("Mean age (Test group):", round(age_test.mean(), 2))
print("Mean age (Control group):", round(age_control.mean(), 2))
print("t-statistic:", round(t_stat, 3))
print("p-value:", round(p_value, 4))

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H. Average age differs between Test and Control groups.")
else:
    print("Conclusion: Fail to reject H. No significant evidence of age difference.")

```

Conclusion:

We compared the average age of clients in the Test vs Control groups using a two-sample t-test. The Test group averaged 47.16 years, while the Control group averaged 47.50 years. The difference of about 0.34 years was statistically significant (p = 0.0157), meaning we reject the null hypothesis. However, given the large sample size, the practical impact of this small age difference may be limited.

What This Means
Statistical significance: Yes, there is evidence that the average age differs between Test and Control groups.

Practical significance: The difference is very small (less than half a year). While statistically real, it may not be meaningful in practice unless even small age shifts matter for your business context.

Interpretation
The average age of clients in the Test group is slightly lower (47.16 vs 47.50).

The difference is small in magnitude (about 0.34 years), but because your sample sizes are very large (tens of thousands of clients), even small differences can become statistically significant.

The negative t-statistic simply reflects the direction: the Test group mean is lower than the Control group mean.
