```python
plt.figure(figsize=(8, 5))

# Use sns.barplot to show the mean of 'bal' (Balance) for each 'gendr' category
sns.barplot(
    data=df_demo,
    x='gendr',
    y='bal',
    palette='viridis'  # A nice color palette
)

plt.title('Average Account Balance by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Account Balance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# AVERAGE balance according to age and genres
plt.figure(figsize=(10, 6))
# Define age ranges: e.g., <30, 30-50, >50
age_bins = [0, 30, 50, 100]
age_labels = ['Young (<=30)', 'Middle (30-50)', 'Senior (>50)']
df_demo['age_group'] = pd.cut(
    df_demo['clnt_age'],
    bins=age_bins,
    labels=age_labels,
    right=False
)
sns.barplot(
    data=df_demo,
    x='age_group',
    y='bal',
    hue='gendr',
    estimator=np.sum, # Corrected: Added comma here
    palette='viridis')

plt.title('Average Account Balance by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Mean Account Balance')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# TOTAL balance according to age and genres
plt.figure(figsize=(10, 6))

sns.barplot(
    data=df_demo,
    x='age_group',
    y='bal',
    hue='gendr',
    estimator=np.sum, # Corrected: Added comma here
    palette='viridis'
)

# Corrected labels for SUM
plt.title('Total Account Balance by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Total Account Balance (Sum)')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
